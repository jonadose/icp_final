```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('screwdriverstatisticsII.csv', sep='\s+', delimiter=';', header = 0, parse_dates=True, squeeze=True, decimal = "," )
df.head()
df.set_index('Id')
```

```python
#Globally Define number of clusters 

n_clusters = 2
```

```python
x = df["RA"]
y = df["T1"]

plt.scatter(x = x, y = y, c = 'b')
plt.show()
```

```python
cluster = df[['RA','T1']]
KMean = KMeans(n_clusters=n_clusters)
KMean.fit(cluster)
cluster_centroids = KMean.cluster_centers_
```

```python
plt.scatter(x = x, y = y, c = 'b')
for i in range(n_clusters): 
    plt.scatter(cluster_centroids[i][0], cluster_centroids[i][1], c = 'r', marker='s')
```
